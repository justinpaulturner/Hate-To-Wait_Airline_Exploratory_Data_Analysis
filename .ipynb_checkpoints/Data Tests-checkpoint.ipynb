## Day Two


```python
### Set environment ###

import numpy as np
import pandas as pd
import seaborn as sns; sns.set(style="ticks", color_codes=True)
import matplotlib.pyplot as plt
%matplotlib inline

### Import January 2019 Flight Data, Airline Names, Airport Names ###

df1 = pd.read_csv('data/1-2019.zip')
airline_names_df = pd.read_csv('data/L_UNIQUE_CARRIERS.csv')
airport_names_df = pd.read_csv('data/Airport Name Codes.csv')

### Show Columns In Imported Flight Dataset ##
for col in df1.columns:
    print(col)
    
### Create Useful Dataframes ###
    
df_consolidated = df1[['DAY_OF_WEEK','ORIGIN',
                       'DEST','DEP_DELAY',
                       'ARR_DELAY',
                       'ARR_DEL15',
                       'CANCELLED',
                       'DIVERTED',
                       'CRS_ELAPSED_TIME',
                       'DISTANCE']]
df_few = df1[['OP_UNIQUE_CARRIER',
              'ORIGIN','DEST',
              'DEP_DELAY',
              'ARR_DELAY',
              'CANCELLED',
              'DIVERTED',
              'CRS_ELAPSED_TIME',
              'DISTANCE']]
merged_df = pd.merge(df_few,airline_names_df, on='OP_UNIQUE_CARRIER')
merged_df = pd.merge(merged_df,airport_names_df, on='ORIGIN')

#OPTIONAL: merge within a merge: merged_df = pd.merge(pd.merge(df_few,airline_names_df, on='OP_UNIQUE_CARRIER'),airport_names_df, on='ORIGIN')


grouped_and_sorted_by_airline_df = merged_df.groupby('Description_x').mean().sort_values(by=['ARR_DELAY'])
#grouped_and_sorted_by_airline_df
grouped_and_sorted_by_airport_df = merged_df.groupby('Description_y').mean().sort_values(by=['DEP_DELAY'])
#grouped_and_sorted_by_airport_df



```

```python
### Plot the airlines by arrival delay, sorted from lowest to highest. Negative values mean airline is early. ###

plt.figure(figsize=(20, 3))
plt.xticks(rotation=15)
plt.bar(grouped_and_sorted_by_airline_df['ARR_DELAY'].index, grouped_and_sorted_by_airline_df['ARR_DELAY'].values)
```

```python
plt.figure(figsize=(100, 33))
plt.xticks(rotation=90)
plt.bar(grouped_and_sorted_by_airport_df['DEP_DELAY'].index, grouped_and_sorted_by_airport_df['DEP_DELAY'].values)
```

```python
grouped_and_sorted_by_airport_df.tail()
```

```python
plt.figure(figsize=(40, 10))
plt.xticks(rotation=90)
plt.bar(grouped_and_sorted_by_airport_df.tail(30)['DEP_DELAY'].index, grouped_and_sorted_by_airport_df.tail(30)['DEP_DELAY'].values)
```

```python
plt.figure(figsize=(40, 10))
plt.xticks(rotation=90)
plt.bar(grouped_and_sorted_by_airport_df.tail(30)['DEP_DELAY'].index, grouped_and_sorted_by_airport_df.tail(30)['DEP_DELAY'].values)
```
